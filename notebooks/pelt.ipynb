```python
import numpy as np
import ruptures as rpt
import matplotlib.pyplot as plt
```

```python
# Fake signal with change points
signal = np.concatenate([
    np.random.normal(0, 1, 150),
    np.random.normal(3, 1, 200),
    np.random.normal(-2, 1, 150),
])

# Fit PELT
model = rpt.Pelt(model="l2").fit(signal)

# Predict change points
penalty = 10
breakpoints = model.predict(pen=penalty)

print("Change points:", breakpoints)
```
